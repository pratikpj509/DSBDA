```python
import pandas as pd #Python library used for working with data sets
import numpy as np #Python library used for working with arrays
import seaborn as sns # library for making statistical graphics in Python
import matplotlib.pyplot as plt #used to create 2D graphs and plots by using python scripts
```

```python
df = pd.read_csv("/home/bharat/DSBDA PRACTICALS/DATASETS/dataset_Facebook.csv",sep=";")
df.head()
```

```python
plt.figure(figsize = (12, 8))
```

```python
# Histogram of Lifetime Engaged Users
# Create a subplot with a single plot (1 row, 1 column, subplot 3)
plt.subplot(1, 1, 1)

# Create a histogram of the 'Lifetime Engaged Users' column
# Set the number of bins to 20 for the histogram
# Set the edgecolor of the bins to black for better visibility
plt.hist(df['Lifetime Engaged Users'], bins=20, edgecolor='k')

# Set the label for the x-axis
plt.xlabel=('Lifetime Engaged Users')

# Set the label for the y-axis
plt.ylabel=('Frequency')

# Set the title of the plot
plt.title=('Lifetime Engaged Users Distribution')

# Display the plot
plt.show()
```

```python
# Create a subplot for the bar plot of Total Interactions
plt.subplot(1, 1, 1)  # Subplot (2 rows, 2 columns, subplot 1)
plt.bar(df['Type'], df['Total Interactions'])  # Bar plot
plt.xlabel=('Type')  # X-axis label
plt.ylabel=('Total Interactions')  # Y-axis label
plt.title=('Total Interactions by Type')  # Plot title

# If you have additional subplots or plot types to add, you can continue to add them here
# For example, you can add a line plot, scatter plot, etc.

# Show the plots
plt.show()
```

```python
# Scatter plot of Lifetime Post Total Reach vs. Lifetime Post Total Impressions
# Create a subplot with a single plot (1 row, 1 column, subplot 1)
plt.subplot(1, 1, 1)

# Create a scatter plot with Lifetime Post Total Reach on the x-axis and Lifetime Post Total Impressions on the y-axis
# Set the transparency of markers to 0.5 for better visualization
plt.scatter(df['Lifetime Post Total Reach'], df['Lifetime Post Total Impressions'], alpha=0.5)

# Set the label for the x-axis
plt.xlabel=('Lifetime Post Total Reach')

# Set the label for the y-axis
plt.ylabel=('Lifetime Post Total Impressions')

# Set the title of the plot
plt.title=('Lifetime Post Total Reach vs. Impressions')

# Display the plot
plt.show()
```

```python
# Box plot of Lifetime Post Consumers by Category
plt.subplot(1, 1, 1)  # Create a subplot (2 rows, 2 columns, subplot 4)
plt.boxplot([df[df['Category'] == 1]['Lifetime Post Consumers'],
             df[df['Category'] == 2]['Lifetime Post Consumers'],
             df[df['Category'] == 3]['Lifetime Post Consumers']],
            labels=['Category 1', 'Category 2', 'Category 3'])
plt.xlabel=('Category')
plt.ylabel=('Lifetime Post Consumers')
plt.title=('Lifetime Post Consumers by Category')

plt.tight_layout()  # Adjust the spacing between subplots
plt.show()  # Display the plots
```

```python
# Create the bar plot using Seaborn
sns.barplot(data=df, x='Type', y='Total Interactions')

# Set labels for x and y axes
plt.xlabel=('Paid')
plt.ylabel=('Lifetime Post Consumers')

# Set the title of the plot
plt.title=('Lifetime Post Consumers by Paid')

# Display the plot
plt.show()
```
