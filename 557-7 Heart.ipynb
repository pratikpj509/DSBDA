```python
import pandas as pd #Python library used for working with data sets
import numpy as np #Python library used for working with arrays
import seaborn as sns # library for making statistical graphics in Python
import matplotlib.pyplot as mat #used to create 2D graphs and plots by using python scripts
```

```python
# Read the CSV file 'heart.csv' into a Pandas DataFrame
df = pd.read_csv('/home/bharat/DSBDA PRACTICALS/DATASETS/heart.csv')
```

```python
import matplotlib.pyplot as plt  # Importing the matplotlib library for plotting

plt.figure(figsize=(10, 6))  # Setting the figure size

plt.subplot(2, 2, 1)  # Creating a subplot (2 rows, 2 columns, subplot 1)
plt.hist(df['age'], bins=10, edgecolor='k')  # Creating a histogram of the 'age' column with 10 bins and black edges
plt.xlabel('Age')  # Labeling the x-axis
plt.ylabel('Frequency')  # Labeling the y-axis
plt.title('Age Distribution')  # Adding a title to the plot

plt.show()  # Displaying the plot
```

```python
# Bar plot of sex
plt.subplot(2, 2, 2)  # Create a subplot (2 rows, 2 columns, subplot 2)

# Count the occurrences of each sex category
sex_counts = df['sex'].value_counts()

# Create a bar plot with the counts of each sex category
plt.bar(sex_counts.index, sex_counts.values)

# Label the x-axis
plt.xlabel('Sex (0 = female, 1 = male)')

# Label the y-axis
plt.ylabel('Count')

# Add a title to the plot
plt.title('Gender Distribution')
```

```python
# Scatter plot of age and cholesterol
plt.subplot(2, 2, 3)  # Create a subplot (2 rows, 2 columns, subplot 3)

# Generate a scatter plot of age against cholesterol with transparency set to 0.5
plt.scatter(df['age'], df['chol'], alpha=0.5)

# Label the x-axis
plt.xlabel('Age')

# Label the y-axis
plt.ylabel('Cholesterol')

# Add a title to the plot
plt.title('Age vs. Cholesterol')
```

```python
# Box plot of resting blood pressure by target
plt.subplot(2, 2, 4)  # Create a subplot (2 rows, 2 columns, subplot 4)

# Create a box plot of resting blood pressure separated by target categories
plt.boxplot([df[df['target'] == 0]['trestbps'],  # Blood pressure values for no disease
             df[df['target'] == 1]['trestbps']],  # Blood pressure values for disease
            labels=['No Disease', 'Disease'])  # Label the box plot categories

# Label the x-axis
plt.xlabel('Target')

# Label the y-axis
plt.ylabel('Resting Blood Pressure')

# Add a title to the plot
plt.title('Resting Blood Pressure by Target')

plt.tight_layout()  # Adjust the spacing between subplots
plt.show()  # Display the plots
```
